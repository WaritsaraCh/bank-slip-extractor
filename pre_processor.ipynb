```python
import cv2
import matplotlib.pyplot as plt

# 
img = cv2.imread(r"static\images\bank_templates\scb_bank.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Preprocess
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

thresh_otsu = cv2.threshold(
    gray, 0, 255, cv2.THRESH_BINARY_INV + cv2.THRESH_OTSU
)[1]

blurred = cv2.GaussianBlur(gray, (3, 3), 0)

adaptive = cv2.adaptiveThreshold(
    blurred,
    255,
    cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
    cv2.THRESH_BINARY,
    11,
    2
)

# Plot
fig, axes = plt.subplots(2, 4, figsize=(16, 8))

# Row 1
axes[0, 0].imshow(img)
axes[0, 0].set_title("Original (RGB)")
axes[0, 0].axis("off")

axes[0, 1].imshow(gray, cmap="gray")
axes[0, 1].set_title("Grayscale")
axes[0, 1].axis("off")

axes[0, 2].imshow(thresh_otsu, cmap="gray")
axes[0, 2].set_title("Otsu Threshold (INV)")
axes[0, 2].axis("off")

axes[0, 3].imshow(blurred, cmap="gray")
axes[0, 3].set_title("Gaussian Blur")
axes[0, 3].axis("off")

# Row 2
axes[1, 0].imshow(adaptive, cmap="gray")
axes[1, 0].set_title("Adaptive Threshold")
axes[1, 0].axis("off")

#  
for i in range(1, 4):
    axes[1, i].axis("off")

plt.tight_layout()
plt.show()
```
